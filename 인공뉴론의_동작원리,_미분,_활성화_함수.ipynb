- 인공뉴론 가중치
  - y=wx
  - y=ax
  - a: 기울기, w: 가중치
- 뉴런의 동작 원리
  - 순입력 : w1x1+w2x2
  - 임계값 : θ theta. -> 임계값을 넘어설 때, 뉴론에서 출력됨
  - 활성화 : θ theta. -> 뉴론의 활성화 여부를 결정하는 한계값
- 인공뉴론 - 가중치
- 편향(bias)
